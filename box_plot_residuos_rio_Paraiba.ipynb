```python
# import packages and libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# reading the csv file
data = pd.read_csv('Paraiba_calib.csv', sep=";", decimal=",", header=0)
```

```python
data.head()
```

```python
# fit simple linear regression model
linear_model = ols('y ~ x', data=data).fit()
```

```python
# display model summary
print('regressão de calibração para o Rio Paraiba do Sul')
print('')
print(linear_model.summary())
```

```python
# modify figure size
fig = plt.figure(figsize=(12, 8))

# creating regression plots
fig = sm.graphics.plot_regress_exog(linear_model, 'x', fig=fig)
```
